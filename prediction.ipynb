```python
# prompt: affiche les colonnes qui contienne des valeurs categorielles

cat_cols = df.select_dtypes(include='object').columns.tolist()
print(cat_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
columns_to_encode = ['Business Travel', 'CF_attrition label', 'Department', 'Education Field', 'Gender', 'Job Role', 'Marital Status', 'Over Time', 'Education']

bt= LabelEncoder()
df['Business Travel'] = bt.fit_transform(df['Business Travel'])
cf = LabelEncoder()
df['CF_attrition label'] = cf.fit_transform(df['CF_attrition label'])
dp = LabelEncoder()
df[ 'Department'] = dp.fit_transform(df[ 'Department'])
ef= LabelEncoder()
df['Education Field'] = ef.fit_transform(df['Education Field' ])
ge= LabelEncoder()
df['Gender' ] = ge.fit_transform(df[ 'Gender'])
jr= LabelEncoder()
df['Job Role' ] = jr.fit_transform(df['Job Role' ])
ms= LabelEncoder()
df['Marital Status'] = ms.fit_transform(df[ 'Marital Status' ])
ot= LabelEncoder()
df['Over Time' ] = ot.fit_transform(df['Over Time' ])
ed= LabelEncoder()
df['Education' ] = ed.fit_transform(df[ 'Education'])
```

```python
# prompt: encode f['Attrition']  1 pour Yes et 0 pour No

att = LabelEncoder()
df['Attrition'] = att.fit_transform(df['Attrition'])
```

```python
df.head()
```

```python
df['CF_age band'] = df['CF_age band'].replace({'Under 25': 18-25, 'Over 55': 55-65})
```

```python
df['CF_age band'].unique()
```

```python
# Convertir les tranches d'Ã¢ge en valeurs numÃ©riques
def convert_age_band_to_numeric(age_band):
    if isinstance(age_band, str) and '-' in age_band:
        age_range = age_band.split('-')
        return (int(age_range[0]) + int(age_range[1])) / 2
    else:
        return age_band

df['CF_age band'] = df['CF_age band'].apply(convert_age_band_to_numeric)

# Afficher le DataFrame mis Ã  jour
print(df)
```

```python
df.head()
```

```python
# prompt: drop the columns "'emp no','Employee Number','Over18',''Employee Count"

df.drop(columns=['emp no','Employee Number','Over18','Employee Count'], inplace=True)
```

```python
df.head()
```

```python
# prompt: split the data

from sklearn.model_selection import train_test_split

# Separate features and target
features = df.drop('Attrition', axis=1)
target = df['Attrition']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)

# Print the shapes of the train and test sets
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', X_test.shape)
print('y_test shape:', y_test.shape)
```

```python

from sklearn.svm import SVC

# Create a new SVM classifier
svm_classifier = SVC(kernel='linear')

# Train the classifier on the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred_svm = svm_classifier.predict(X_test)

# Evaluate the performance of the classifier using accuracy
from sklearn.metrics import accuracy_score
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accuracy_svm)
```

```python
from xgboost import XGBClassifier

# Create an XGBClassifier object
xgb_classifier = XGBClassifier()

# Train the classifier on the training data
xgb_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred_xgb = xgb_classifier.predict(X_test)

# Evaluate the performance of the classifier using accuracy
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print("XGBoost Accuracy:", accuracy_xgb)
```
